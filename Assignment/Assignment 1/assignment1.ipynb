```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
from google.colab.patches import cv2_imshow  # Import for displaying in Colab
```

```python
def extract_center_region(image_url, size=(100, 100)):
  resp = urllib.request.urlopen(image_url)
  image = np.asarray(bytearray(resp.read()), dtype="uint8")
  image = cv2.imdecode(image, cv2.IMREAD_COLOR)
  h, w = image.shape[:2]
  x = w//2 - size[0]//2
  y = h//2 - size[1]//2
  return image[y:y+size[1], x:x+size[0]]

# Example usage
image_url = "https://static1.colliderimages.com/wordpress/wp-content/uploads/2021/11/Luffy-One-Piece-Character.jpeg"
center_region = extract_center_region(image_url)
cv2_imshow(center_region)
```

```python
def visualize_hsv_components(image_url):
  resp = urllib.request.urlopen(image_url)
  image = np.asarray(bytearray(resp.read()), dtype="uint8")
  image = cv2.imdecode(image, cv2.IMREAD_COLOR)
  hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
  h, s, v = cv2.split(hsv_image)
  cv2_imshow(h)
  cv2_imshow(s)
  cv2_imshow(v)

# Example usage
image_url = "https://static1.colliderimages.com/wordpress/wp-content/uploads/2021/11/Luffy-One-Piece-Character.jpeg"
visualize_hsv_components(image_url)
```
